```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
#reading the csv file

df = pd.read_csv('Book.csv')
df.head()
```

```python
# dropping columns

drop_columns = ['Edition Statement', 'Author','Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver', 'Issuance type',
               'Flickr URL', 'Shelfmarks']

df.drop(drop_columns, inplace = True, axis = 1)
df.head()
```

```python
remove_characters = ['[', ',', '-','?','.']

def clean_date(item):
    dop= str(item.loc['Date of Publication'])
                     
    if dop == 'nan' or dop[0] == '[':
        return np.NaN
    
    for character in remove_characters:
        if character in dop:
            character_index = dop.find(character)
            dop = dop[:character_index]
                     
    return dop

df['Date of Publication'] = df.apply(clean_date, axis = 1)
```

```python
df.head()
```

```python
def clean_pub_place(item):
    place = str(item['Place of Publication'])
    
    if 'London' in place:
        return 'London'
    if 'Oxford' in place:
        return 'Oxford'
    if place == 'Newcastle upon Tyne':
        return 'Newcastle-upon-Tyne'
    
    return place
    
df['Place of Publication'] = df.apply(clean_pub_place, axis = 1)
```

```python
df.head()
```

```python
def clean_title(item):
    title = str(item['Title'])
    
    if title == 'nan':
        return np.NaN
    
    if title[0] == '[':
        title = title[1: title.find(']')]
        
    if 'by' in title:
        title = title[:title.find('by')]
    elif 'By' in title:
        title = title[:title.find('By')]
        
    if '[' in title:
        title = title[:title.find('[')]

    title = title[:-2]
        
    title = list(map(str.capitalize, title.split()))
    return ' '.join(title)
    
df['Title'] = df.apply(clean_title, axis = 1)
```
